# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
@app.route('/get-profile-url/<username>')
def get_profile_url(username):
    # Using url_for to build a URL with a dynamic segment
    profile_url = url_for('profile', username=username)
    return f'The URL for {username}\'s profile is {profile_url}'
```
