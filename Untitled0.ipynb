**Q5. What is inheritance? Give an example for each type of inheritance.**

```python
Inheritance="""
          when we define a class that class inharit proparties of other class is called
          inheritance.
"""
```

```python
Single_inheritance=' A class inherits from only one parent class.'
```

```python
class Father:
  def Money(self):
    print(' Bank balance 50 lakhs')

# inharite father class

class Child(Father):
  def pocket_money(self):
    print(' Pocket money 50 rupees')

# object
child=Child()
child.Money()
```

```python
Multiple_inheritance=' A class inherits from multiple parent classes. '
```

```python
class Father:
  def Money(self):
    print(' Bank balance 50 lakhs')

class Mother:
  def money(self):
    print('bank balance 10 lakhs')

# inharite father class

class Child(Father,Mother):
  def pocket_money(self):
    print(' Pocket money 50 rupees')

# object
child=Child()
child.Money()
child.money()
```

```python
Hierarchical_inheritance= ' A class inherits from a parent class,a parent and multipel child  calss but each child class can access paren calss proparty .'
```

```python
class Father:
  lastname='chakraborty'
  def last_name(self):
    print(f'My last name : {self.lastname}')

class child1(Father):
  def son(self):
    print(f"my name is arapn {self.lastname}")

class child2(Father):
  def daughter(self):
    print(f"my name is aprajita {self.lastname}")
```

```python
Multilevel_inheritance='''
              A class inherits from a parent class,
              which inherits from another parent class, and so on,
              with only one level of inheritance between each class'''
```

```python
class data:
  a=20
  b=30

class multiply(data):
  def multiple(self):
    return (self.a  * self.b)

class sub(multiply):
  def subtract(self):
    return (self.a,self.b,self.b -self.a)
```

```python
obj=multiply()
obj.multiple()
```

```python
obj=sub()
obj.subtract()
```
